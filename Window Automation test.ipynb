```python
import numpy as np
import matplotlib.pyplot as plt
from hub_tools import *
import os
```

```python
def edge(image):
    return np.concatenate((image[0,:-1],image[:-1,-1],image[-1,-1:0:-1],image[-1:0:-1,0]))

#def saturation_center(image):
    
    

def adjust_window(image,square=False,shape=None):
    
    cut_image = image.copy()
    image_edge = edge(cut_image)
    while np.isnan(image_edge).any():
        cut_image = cut_image[1:-1,1:-1]
        image_edge = edge(cut_image)
    
    return cut_image
```

```python
arr = np.array([[1,2,1,np.nan],
                [3,2,4,1],
                [2,3,4,1],
                [np.nan,0,2,3]],float)
adjust_window(arr)
```
