```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astropy.io import fits
from hub_simulation import fit
```

```python
df = pd.read_table('sat_sources_data/selection_FWHM.txt')
df = df.sort_values(by='L')
df.index = range(len(df))
print(df)

files = os.listdir('../selected')
files.sort()
files = files[1:8*3+1]
files
```

```python
hdulist = fits.open('../selected/'+files[0])
hdu = hdulist[0]
header = hdu.header
data = hdu.data

x = header['crval1'] + header['cdelt1']*np.arange(0,header['naxis1'],1)
y = header['crval2'] + header['cdelt2']*np.arange(0,header['naxis2'],1)
y = y[::-1]

x *= 3600
y *= 3600

xl,xr = 43.55*3600,43.48*3600
yb,yt = 0.42*3600,0.46*3600

X,Y = np.meshgrid(x,y)

data_sub = data[(y>=yb)&(y<=yt),:][:,(x<=xl)&(x>=xr)]

plt.figure(figsize=(8,8))
plt.imshow(np.log10(data),extent=(x[0],x[-1],y[-1],y[0]))
plt.show()

plt.figure(figsize=(8,8))
plt.imshow(np.log10(data_sub),extent=(xl,xr,yb,yt))
plt.show()

n = 5
grid = np.meshgrid(x[(x<=xl)&(x>=xr)],y[(y>=yb)&(y<=yt)])
sat_area = np.isnan(data_sub)
params,corrected = fit(grid,data_sub,sat_area,FWHM=3*np.array(n*[[28.276492,28.276492]]),peaks=n)
plt.figure(figsize=(8,8))
plt.imshow(np.log10(corrected),extent=(xl,xr,yb,yt))
for i in range(n):
    plt.plot(params[i*6],params[i*6+1],'ro')
plt.colorbar()
plt.show()
```
